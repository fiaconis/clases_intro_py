### ¿Qué es una variable?
Una variable es como una bolsa. Dentro de esa bolsa podemos poner muchas cosas diferentes, desde una botella de gaseosa, hasta un caramelo. En este caso lo que vamos a poner dentro de la bolsa es nuestro nombre.

    bolsa = "Pancho"
    
En este caso la variable es 'bolsa' que guarda dentro mi nombre. Detalle importante: a la compu hay que indicarle que lo que guardamos es texto con comillas, sino va a interpretar que pancho es otra variable/bolsa.

Las variables pueden llamarse como ustedes quieran, pero hay algunas pocas restricciones. Las variables no pueden ser números y tampoco se pueden llamar como algún comando de python (esto lo vamos a ir viendo ams adelante).

Las variables pueden ser basicamente de tres tipos diferentes ( hay mas, pero son combinaciones de estas tres): 
- **Texto**, en ingles string, o su manera abreviada *str*
- **Número entero** (aquellos numeros que no tienen coma), en ingles integer, o su manera abreviada *int*
- **Número decimal** (aquellos numeros que tienen coma, por ejemplo 2,4). A estos numeros se los denomina "numeros de punto flotante", que en ingles es *float*


Una variable se crea en el momento que le asignamos un valor. y el valor que tiene es el ultimo asignado

    x = 4 # x es una variable int, un cuatro
    x = 'Pepe' # x ahora es un string y su valor es 'Pepe'
    print(x)
    Pepe

Podemos especificar el tipo de variable 'Casteando' (la españolizacion del verbo 'Casting').

    x = str(3)    # x va a ser '3', un string.
    y = int(3)    # y va a ser 3, un entero, o integer, int.
    z = float(3)  # z va a ser 3.0, un numero con decimal, un float.

### ¿De qué manera le digo mi nombre a la compu?

Hay varias maneras en la que podemos pasarle informacion a la compu. Una de esas maneras es como hicimos mas arriba asignandole mi nombre a la variable 'bolsa'. Otra manera es utilizando el comando 'input()':

```python
nombre = input('Ingrese su nombre: ')
```

El comando 'input' crea un cuadro de dialogo para que el usuario ingrese por teclado la información. Cualquier cosa que vayamos a poner ahi dentro se va a guardar en la variable 'nombre' **como un string, no importa si le ponemos un número**. Esta variable va a ser una variable que va a considerar cualquier cosa que le pongamos como un texto, es decir como un string.

Otra manera de ingresar informacion a la computadora la vamos a ver mas adelante pero les adelanto que es pidiendole a la compu que lea el contenido de archivos de texto.

### ¿Cómo hago para que la compu me salude?

Tenemos que tener alguna manera de comunicarnos con la computadora. Hasta ahora nosotros "le dijimos" nuestro nombre. Haora tenemos que hacer que la compu nos devuelva un saludo. Para esto utilizamos el comando "print" (imprimir en ingles)

```python
print(nombre)
```

Fijense que le dijimos que imprima el contenido de la "bolsa"/variable 'nombre'. Dentro de esa variable estaba la palabra "Pancho" y lo que hizo fue volcar el contenido de la bolsa en la pantalla para que podamos leerla.

Pero eso no es un saludo, eso es simplemente mi nombre! hay que agregarle algo mas ese print. a continuacion vamos a ver tres maneras de imprimir un saludo:

```python
print('Hola,',nombre,'!') # imprime un texto (string), seguido de una variable y seguido de otro string) 
print('Hola, '+nombre+'!') # imprime concatenacion de textos (strings)
print(f'Hola, {nombre}!') # imprime texto con formato
print('Hola, {}!'.format(nombre)) #otra manera de imprimir con formato
```

Como veran, hay varias formas de que la compu nos muestre informacion. Utilicen la forma que mas les guste.
