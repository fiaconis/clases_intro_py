```python
def fun(s): # Voy a crear una funcion
    for i in range(len(s)): # que recorra todo el archivo caracter por caracter
        #y cuando encuentre un "\n" y que el caracter siguiente NO sea un espacio y que ademas el texto siguiente no sea un def entonces
        if (s[i] == '\n') and (s[i+1] != ' ') and (s[i+1:i+4] != 'def') :
            s = s[:i+1]+'    '+s[i+1:] # introduzca un '     ' entre el "\n" y el texto que sigue
    return(s)

new_codigo_1 = fun(codigo) 
print(new_codigo_1)
```
